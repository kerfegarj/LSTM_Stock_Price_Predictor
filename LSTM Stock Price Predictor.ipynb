```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Neural Networks/TeslaStockPrice.csv')
df
```

```python
table=[]
row=[]
for i in range(0,len(df)-10):
  for j in range(i,i+10):
    row.append(df["Close"][j])
  row=[]
  table.append(row)
```

```python
table=pd.DataFrame(table)
table=table.dropna()
table.reset_index()
table
```

```python
table["mean"]=table.mean(axis=1)
```

```python
table
```

```python
plt.plot(table["mean"])
plt.xticks()
plt.yticks()
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
x=table.drop([9,"mean"],axis=1)
y=table[9]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xteat,ytrain,ytest=train_test_split(x,y,train_size=0.2)
```

```python
model=Sequential()
model.add(LSTM(100,input_shape=(9,1)))
model.add(Dense(32,activation="relu"))
model.add(Dense(16,activation="relu"))
model.add(Dense(1))
model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
model.summary()
```

```python
model.fit(xtrain,ytrain,epochs=50)
```

```python
ypreds=model.predict(x)
ypreds=np.reshape(ypreds,(1681))
```

```python
comp=pd.DataFrame({"Actual":y,"Predicted":ypreds})
comp
```

```python
plt.scatter(comp["Actual"],comp["Predicted"])
plt.plot([0,300],[0,300],color="red")
plt.show()
```
